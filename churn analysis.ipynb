```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('customer churn.csv')
df.describe()


```

```python
#  Churn rate

churn_rate = df["Churn Label"].value_counts(normalize=True) * 100
print("Churn Rate:\n", churn_rate)
```

```python
# visualization

plt.figure(figsize=(4,4))
sns.countplot(data=df, x="Churn Label", hue="Gender")
plt.title("Churn by Gender")
plt.show()

plt.figure(figsize=(4,4))
sns.boxplot(data=df, x="Churn Label", y="Monthly Charge")
plt.title("Monthly Charges vs Churn")
plt.show()

```

```python
#  Churn by Contract & Payment

plt.figure(figsize=(5,4))
sns.countplot(data=df, x="Contract", hue="Churn Label")
plt.title("Churn by Contract Type")
plt.xticks(rotation=20)
plt.show()

plt.figure(figsize=(5,4))
sns.countplot(data=df, x="Payment Method", hue="Churn Label")
plt.title("Churn by Payment Method")
plt.xticks(rotation=20)
plt.show()
```

```python
#   churn rate
churn_rate = df["Churn Label"].value_counts(normalize=True) * 100
print("Churn Rate (%)\n", churn_rate)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="Churn Label", data=df)
plt.title("Overall Churn Distribution")
plt.show()
```

```python
senior_churn = df.groupby("SeniorCitizen")["Churn Label"].value_counts(normalize=True).unstack() * 100
print("Churn by Senior Citizen:\n", senior_churn)
```
